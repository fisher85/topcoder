```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
y_train = train["target"]
X_train = train.drop(["target"], 1)
X_test = test
```

```python
from sklearn.utils import shuffle
X, y = shuffle(X_train, y_train, random_state=13)
offset = int(X.shape[0] * 0.9)
# X_train, y_train = X[:offset], y[:offset]
# X_test, y_test = X[offset:], y[offset:]
```

```python
from sklearn import ensemble
params = {'n_estimators': 100, 'max_depth': 3, 'min_samples_split': 2,
          'learning_rate': 0.1, 'loss': 'ls'} # 1.29 балла
params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'} # 1.17 балла
params = {'n_estimators': 500, 'max_depth': 3, 'min_samples_split': 2,
          'learning_rate': 0.1, 'loss': 'ls'} # 0 балла - ?
params = {'n_estimators': 200, 'max_depth': 3, 'min_samples_split': 2,
          'learning_rate': 0.1, 'loss': 'ls'} # 2.87 балла
params = {'n_estimators': 300, 'max_depth': 3, 'min_samples_split': 2,
          'learning_rate': 0.1, 'loss': 'ls'} # 0 балла - ?
params = {'n_estimators': 250, 'max_depth': 3, 'min_samples_split': 2,
          'learning_rate': 0.1, 'loss': 'ls'} # 0.08 балла - ?
clf = ensemble.GradientBoostingRegressor(**params)

clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = clf.predict(X_test)
# mse = mean_squared_error(y_test, y_pred)
# print("MSE: %.4f" % mse)
```

```python
prediction = test[["sample_id"]]
prediction["target"] = y_pred
prediction
```

```python
prediction.to_csv("prediction.csv", index=False)
```
